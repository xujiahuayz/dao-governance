```python
%%stata
clear all
set more off
set varabbrev off

global PROCESSED_DATA "processed_data"
global TABLES "tables"

import delimited using "$PROCESSED_DATA/proposals_panel.csv", clear

* Dependent variables
local cars car_created car_end
local wins non_whale_victory_vn non_whale_victory_vp non_whale_victory_vp_vn
local proposal_char weighted ranked_choice multi_choices 
local discussion_char concensus professionalism objectiveness data_intensity hhi_post_number
local topic_char protocol_security treasury_expenditure user_incentive_increase tokenomics voting_mechanism_change

* Label variables
label var car_created "\${\it CAR}[-5,5]_{i}^{\it created}\$"
label var car_end "\${\it CAR}[-5,5]_{i}^{\it end}\$"
label var non_whale_victory_vn "\${\it Small Win}^{vn}_{i}\$"
label var non_whale_victory_vp "\${\it Small Win}^{vp}_{i}\$"
label var non_whale_victory_vp_vn "\${\it Small Win}^{vp,vn}_{i}\$"
label var weighted "\${\it Weighted}_{i}\$"
label var quadratic "\${\it Quadratic Voting}_{i}\$"
label var ranked_choice "\${\it Ranked Choice}_{i}\$"
label var multi_choices "\${\it Multiple Choices}_{i}\$"
label var professionalism  "\${\it Professionalism}_{i}\$"
label var objectiveness    "\${\it Objectiveness}_{i}\$"
label var concensus       "\${\it Concensus}_{i}\$"
label var data_intensity   "\${\it Data Intensity}_{i}\$"
label var hhi_post_number  "\${\it HHI}_{i}^{\# Post}\$"
label var user_incentive_increase "\${\it User Incentive Increase}_{i}\$"
label var treasury_expenditure "\${\it Treasury Expenditure}_{i}\$"
label var voting_mechanism_change "\${\it Voting Mechanism Change}_{i}\$"
label var protocol_security "\${\it Protocol Security}_{i}\$"
label var tokenomics "\${\it Tokenomics}_{i}\$"
```

```python
%%stata

******************************************************
* SUMMARY STATISTICS
******************************************************
* Summarize all key variables
summarize `cars' `wins' `voting' `proposal_char' `discussion_char' `topic_char', detail

* Export a nice LaTeX summary table with mean, SD, p10, median, p90
estpost summarize `cars' `wins' `voting' `proposal_char' `topic_char' `discussion_char', detail
esttab using "$TABLES/summary_stats.tex", replace ///
    cells("count(fmt(0)) mean(fmt(3)) sd(fmt(3)) min(fmt(3)) p25(fmt(3)) p50(fmt(3)) p75(fmt(3)) max(fmt(3))") ///
    collabels("Number of Obs." "Mean" "Std. Dev." "Min" "P25" "P50" "P75" "Max") ///
    nonumber nomtitles noobs label ///
    alignment(D{.}{.}{-1}) booktabs ///
    compress nofloat nostar ///
    substitute("\_" "_")                     ///
    title("Summary Statistics of Variables") ///
    prehead("\begin{tabular}{l*{9}{c}}\toprule") ///
    posthead("\hline") ///
    postfoot("\bottomrule\end{tabular}")
```
