```python
import stata_setup

stata_setup.config('/Applications/StataNow', 'mp')
```

```python
%%stata
clear all
set more off
set varabbrev off

global PROCESSED_DATA "processed_data"
global TABLES "tables"

****************************************************
* Import & sample restriction
****************************************************
import delimited "$PROCESSED_DATA/proposals_panel.csv", clear varn(1) case(lower)
keep if switch_delegation == 0

****************************************************
* Mean comparison & t-test
****************************************************
estpost ttest whale_participation, by(delegation)

* Extract means and compute difference (Enabled - Disabled)
matrix M = r(table)
scalar mean_disabled  = M[1,1]   // mu_1
scalar mean_enabled   = M[1,2]   // mu_2
scalar diff           = mean_enabled - mean_disabled

* Flip t-stat to match the same difference direction
scalar t2 = -M[3,1]

* Add to estpost results
estadd scalar diff
estadd scalar t2

****************************************************
* Export LaTeX
****************************************************
esttab . using "$TABLES/sum_delegation.tex", replace ///
    cells("mu_1(fmt(3)) mu_2(fmt(3)) b(fmt(3)) t(fmt(2)star)") ///
    collabels("Delegation Disabled" "Delegation Enabled" "Disabled - Enabled" "t-stat") ///
    nomtitle nonumber noobs label ///
    varlabels("whale_participation" "\${\it Participation}^{Whale}_{i,t}\$")
```
