```python
%%stata
clear all
set more off
set varabbrev off

global PROCESSED_DATA "processed_data"
global TABLES "tables"

****************************************************
* Import & sample restriction
****************************************************
import delimited "$PROCESSED_DATA/proposals_panel.csv", clear varn(1) case(lower)

****************************************************
* Mean comparison & t-test for multiple variables
* All variables share the same columns
****************************************************

* Put all variables you want here:
local vars protocol_security treasury_expenditure user_incentive_increase tokenomics voting_mechanism_change concensus professionalism objectiveness data_intensity hhi_post_number

* Single estpost with varlist -> shared columns in esttab
estpost ttest `vars', by(infrastructure)

* Export (shared columns; one row per variable)
esttab using "$TABLES/sum_infrastructure.tex", ///
    cells("mu_1(fmt(3)) mu_2(fmt(3)) b(fmt(3)) t(fmt(2)star)") ///
    wide unstack ///
    nonumber ///
    star(* 0.1 ** 0.05 *** 0.01) ///
    label noobs nonumber nomtitles booktabs replace ///
    substitute("\_" "_") ///
    collabels("Without Infrastructure" "With Infrastructure" "Without - With" "t-stat") ///
    varlabels( ///
    protocol_security "\${\it Protocol Security}_{i}\$" ///
    treasury_expenditure "\${\it Treasury Expenditure}_{i}\$" ///
    user_incentive_increase "\${\it User Incentive Increase}_{i}\$" ///
    tokenomics "\${\it Tokenomics}_{i}\$" ///
    voting_mechanism_change "\${\it Voting Mechanism Change}_{i}\$" ///
    concensus "\${\it Concensus}_{i}\$" ///
    professionalism "\${\it Professionalism}_{i}\$" ///
    objectiveness "\${\it Objectiveness}_{i}\$" ///
    data_intensity "\${\it Data Intensity}_{i}\$" ///
    hhi_post_number "\${\it HHI}_{i}^{\# Post}\$" ///
    )
```
